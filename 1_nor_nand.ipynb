Wykorzystując podstawowy model sztucznego neuronu należy zaimplementować bramki logiczne.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

```python
def neuron(X, W, threshold=0):
    net = np.dot(X, W)
    return 1 if net >= threshold else 0
```

```python
X1, X2 = zip(*product((0, 1), repeat=2))
```

```python
NOR_result = []
NAND_result = []
for x1, x2 in zip(X1, X2):
    nor = neuron(neuron([x1, x2], [1, 1], 1), -1)
    NOR_result.append(nor)
    print('---NOR---')
    print(f'{x1}, {x2} -> {nor}')
    nand = neuron([neuron(x1, -1), neuron(x2, -1)], [1, 1], 1)
    NAND_result.append(nand)
    print('---NAND---')
    print(f'{x1}, {x2} -> {nand}')
```

```python
plt.figure(figsize=(12, 9))
colormap = np.array(['r', 'g'])
categories = NOR_result
plt.scatter(X1, X2, c=colormap[categories], s=550, marker="o")
```

```python
plt.figure(figsize=(12, 9))
colormap = np.array(['r', 'g'])
categories = NAND_result
plt.scatter(X1, X2, c=colormap[categories], s=550, marker="o")
```
